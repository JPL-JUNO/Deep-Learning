```python
text = 'You say goodbye and I say hello.'
```

```python
text = text.lower()
# 考虑到句子结尾处的句号（.），我们先在句号前插入一个空格（即用“ .”替换“.”）
# 这只是一种临时对策，可以使用正则表达式来进行分词
text = text.replace('.', ' .')
text
```

```python
words = text.split(' ')
words
```

```python
word_to_id = {}
id_to_word = {}
for word in words:
    if word not in word_to_id:
        new_id = len(word_to_id)
        word_to_id[word] = new_id
        id_to_word[new_id] = word
        
# for word in words:
#     word_to_id.setdefault(word, len(word_to_id))
# 不能确定这种写法是否更高效，但是上面一行效率更高
# id_to_word = {v: k for k, v in word_to_id.items()}
```
